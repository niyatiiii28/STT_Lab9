```python
import mlrun

# Create new project
project = mlrun.new_project(
    "breast-cancer-pipeline",
    context="./",
    init_git=True
)

# Register functions
project.set_function("dataprep.py", name="data-prep", kind="job", handler="data_prep")
project.set_function("trainer.py", name="train-model", kind="job", handler="train_model")
project.set_function("serving.py", name="serving", kind="serving")

# Set workflow
project.set_workflow("main", "workflow.py")

# Run the workflow
run = project.run(
    "main",
    watch=True
)

# Display results
print(f"Pipeline completed! Run ID: {run.run_id}")
```
